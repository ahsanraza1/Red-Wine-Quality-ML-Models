```python
print("STARTING THE ALGORITHM")
print( "Importing Libraries.......")
print()

import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, recall_score, precision_score, accuracy_score

print("Reading CSV DATASET")

data = pd.read_csv('/kaggle/input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')

print("Separating Target Data")

y = data['quality']
x = data.loc[:, ~data.columns.isin(['quality'])]

print("Splitting Train and Test Data.....")

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=55)

print("Creating DecisionTreeClassifier Model....")

model = DecisionTreeClassifier()
# model = DecisionTreeRegressor()
print("Fitting the Model....")

model.fit( train_x, train_y)

print("Prediction usiong Training Data")

prediction_trained = model.predict( train_x )

print("Initializing Performance metrices...(for Training session)")

training_accuracy_score  = accuracy_score(train_y, prediction_trained)
training_f1_score        = f1_score( train_y, prediction_trained, average = "weighted")
training_recall_score    = recall_score(train_y, prediction_trained, average = "weighted")
training_precision_score = precision_score( train_y, prediction_trained, average = "weighted")

print("Prediction usiong Test Data")

prediction_testing = model.predict( test_x )

print("Initializing Performance metrices...(for Testing session)")

test_accuracy_score  = accuracy_score(test_y, prediction_testing )
test_f1_score        = f1_score( test_y, prediction_testing, average = "weighted" )
test_recall_score    = recall_score(test_y, prediction_testing, average = "weighted" )
test_precision_score = precision_score( test_y, prediction_testing, average = "weighted" )

print("")
print("Performance metrices...(for Training session)")
print("")

print( "Training Accuracy Score : ", training_accuracy_score)
print( "Training F1 Score       : ", training_f1_score)
print( "Training Recall Score   : ", training_recall_score)
print( "Training Precision Score: ", training_precision_score)

print("")
print("Performance metrices...(for Testing session)")
print("")

print( "Testing Accuracy Score : ", test_accuracy_score)
print( "Testing F1 Score       : ", test_f1_score)
print( "Testing Recall Score   : ", test_recall_score)
print( "Testing Precision Score: ", test_precision_score)
```
